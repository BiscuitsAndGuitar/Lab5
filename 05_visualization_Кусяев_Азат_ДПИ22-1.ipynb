1\. В файле `average_ratings.npy` содержится информация о среднем рейтинге 3 рецептов за период с 01.01.2019 по 30.12.2021. При помощи пакета `matplotlib` в _одной системе координат_ (на одной картинке) изобразите три временных ряда, соответствующих средним рейтингам этих рецептов. 

По горизонтальной оси располагается номер дня (0, 1, 2, ...), по вертикальной - средний рейтинг рецептов в этот день. 

Названия рецептов и их индексы в файле `average_ratings.npy`:
```
0: waffle iron french toast
1: zwetschgenkuchen bavarian plum cake
2: lime tea
```

Результатом работы является визуализация, на которой:
* добавлена подпись горизонтальной оси с текстом "Номер дня"
* добавлена подпись вертикальной оси с текстом "Средний рейтинг"
* добавлена подпись рисунка с текстом "Изменение среднего рейтинга трех рецептов"
* каждый из временных рядов имеет уникальный цвет
* добавлена легенда, на которой отображается название каждого из рецептов

_Примечание_ : для считывания файла воспользуйтесь функцией `np.load`.

````python
average_ratings = np.load('data/average_ratings.npy')
average_ratings
````

````python
days = [i for i in range(1, len(average_ratings[0]) + 1)]
days
````

````python
fig_1, ax_1 = plt.subplots(figsize=(20, 6))

recipe_1, = ax_1.plot(days, average_ratings[0], 'r', label='waffle iron french toast')
recipe_2, = ax_1.plot(days, average_ratings[1], 'g', label='zwetschgenkuchen bavarian plum cake')
recipe_3, = ax_1.plot(days, average_ratings[2], 'b', label='lime tea')

ax_1.legend(handles=[recipe_1, recipe_2, recipe_3])
ax_1.set_xlabel('Номер дня')
ax_1.set_ylabel('Средний рейтинг')
ax_1.set_title('Изменение среднего рейтинга трех рецептов')
````

3\. Измените визуализацию, полученную в задании 2, разбив одну картинку на три, расположенных друг под другом. Три изображения должны иметь одну общую горизонтальную ось (каждое изображение засечки в нижней части, но значения этих засечек находятся только под самым нижним изображением).
````python
fig_3, (ax_1, ax_2, ax_3) = plt.subplots(nrows=3, ncols=1, figsize=(17, 18), layout='constrained', sharex='col')

recipe_1, = ax_1.plot(dates, average_ratings[0], 'r', label='waffle iron french toast')
recipe_2, = ax_2.plot(dates, average_ratings[1], 'g', label='zwetschgenkuchen bavarian plum cake')
recipe_3, = ax_3.plot(dates, average_ratings[2], 'b', label='lime tea')

axes = [ax_1, ax_2, ax_3]
for ax in axes:
    ax.legend()
    ax.set_ylabel('Средний рейтинг')
    ax.set_ylim([0, 11])
ax_1.set_title('Изменение среднего рейтинга трех рецептов')
ax_3.set_xlabel('Дата')
ax_3.xaxis.set_major_locator(mdates.YearLocator())
ax_3.xaxis.set_minor_locator(mdates.MonthLocator())
````
